```python
import matplotlib.pyplot as plt
import numpy as np

goods = [
    "Total",
    "Agriculture, Hunting, Forestry, and Fishing",
    "Coal and Natural Gas",
    "Mining and Quarrying",
    "Food and Beverages",
    "Textiles and Leather",
    "Wood Products",
    "Petroleum Products",
    "Chemicals and Rubber",
    "Non-Metallic Minerals",
    "Basic Metals",
    "Machinery and Equipment",
    "Transport Equipment",
    "Furniture",
    "Secondary Raw Materials",
    "Mail and Parcels",
    "Transport of Goods",
    "Non-Market Goods",
    "Grouped Goods",
    "Unidentifiable Goods",
]

percentages = [
    0.633062133,
9.38638826,
68.18181818,
2.839596192,
3.991635353,
8.211382114,
3.392872047,
-5.51758794,
-1.315729431,
-2.390758555,
-17.05213342,
3.153750195,
-16.63609882,
-9.516380655,
-13.06073542,
18.9902614,
10.32376809,
-7.00525394,
-2.808534834,
7.613843352,

]

positive_percentages = [x if x > 0 else 0 for x in percentages]
negative_percentages = [x if x < 0 else 0 for x in percentages]

fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.4
index = np.arange(len(goods))

ax.bar(index, positive_percentages, bar_width, color='lightgreen', label='Positive',edgecolor='black')
ax.bar(index, negative_percentages, bar_width, color='lightcoral', label='Negative',edgecolor='black')

for i, (pos, neg) in enumerate(zip(positive_percentages, negative_percentages)):
    if pos > 0:
        ax.text(i, pos + 2, f'{pos:.2f}%', ha='center', va='bottom', fontsize=10, color='green')
    if neg < 0:
        ax.text(i, neg - 2, f'{neg:.2f}%', ha='center', va='top', fontsize=10, color='red')

ax.set_xticks(index)
ax.set_xticklabels(goods, rotation=90, fontsize=10)

ax.set_ylim(top=80, bottom=-45)  

plt.ylabel('Percentage Change')
plt.title('Percentage Change of Goods (2018-2020)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```
