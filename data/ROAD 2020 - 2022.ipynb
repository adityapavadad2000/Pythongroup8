```python
import matplotlib.pyplot as plt
import numpy as np

goods = [
    "Total",
    "Agriculture, Hunting, Forestry, and Fishing",
    "Coal and Natural Gas",
    "Mining and Quarrying",
    "Food and Beverages",
    "Textiles and Leather",
    "Wood Products",
    "Petroleum Products",
    "Chemicals and Rubber",
    "Non-Metallic Minerals",
    "Basic Metals",
    "Machinery and Equipment",
    "Transport Equipment",
    "Furniture",
    "Secondary Raw Materials",
    "Mail and Parcels",
    "Transport of Goods",
    "Non-Market Goods",
    "Grouped Goods",
    "Unidentifiable Goods",
]

percentages_2020_2022 = [
  1.551515577,
-10.2752266,
-2.316602317,
-7.549581577,
9.992064287,
-3.531179564,
-6.550958628,
-7.243910222,
-3.965905238,
9.792928052,
1.200883408,
9.485658138,
-13.89931409,
-0.517241379,
-7.439630172,
40.27568383,
10.65696014,
-16.57250471,
8.756831653,
-3.94380501,



]

# Use a single bar plot and adjust colors based on the sign of the percentage
colors = ['lightgreen' if x > 0 else 'lightcoral' for x in percentages_2020_2022]

fig, ax = plt.subplots(figsize=(16, 10))  # Increase the figsize

bar_width = 0.4
index = np.arange(len(goods))

ax.bar(index, percentages_2020_2022, bar_width, color=colors, edgecolor='black')

for i, perc in enumerate(percentages_2020_2022):
    label_position = np.sign(perc) * (abs(perc) + 5)  # Adjust the value for better separation
    ax.text(i, label_position, f'{perc:.2f}%', ha='center', va='center', fontsize=12, color='black')  # Increase fontsize

ax.set_xticks(index)
ax.set_xticklabels(goods, rotation=90, fontsize=12)  # Increase fontsize

# Let Matplotlib automatically adjust the y-axis limits
ax.autoscale(enable=True, axis='y', tight=True)


ax.set_ylim(top=55, bottom=-50)

plt.ylabel('Percentage Change', fontsize=14)  # Increase fontsize
plt.title('Percentage Change of Goods (2020-2022)', fontsize=16)  # Increase fontsize
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```
